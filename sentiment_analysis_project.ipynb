4.Remove Stop Word

remove stopword-some common word that are present in the text but do not contribute in meaning of sentence

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download necessary resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Download the missing 'punkt_tab' resource

def remove_stopwords(text):
    # Set of English stop words
    stop_words = set(stopwords.words('english'))

    # Tokenize the input text
    words = word_tokenize(text)

    # Filter out stop words
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Return filtered words (can also return as string if needed)
    return filtered_words
```

```python
df['review']=df['review'].apply(remove_stopwords)
```

5. **Stemming**

Stemming is a process in Natural Language Processing (NLP) where words are reduced to their root form (or stem), typically by removing suffixes or prefixes.

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
def porter_stemming(words):
    return [stemmer.stem(word) for word in words]
```

Feature Engg.(Text vecterizer)

A Text Vectorizer is a tool used in Natural Language Processing (NLP) to convert text data into numerical format so machine learning models can understand and process i

using TF-IDF

TF-IDF stands for Term FrequencyInverse Document Frequency. It's a text vectorization technique used to convert text into meaningful numerical vectors.

It measures how important a word is in a document relative to a collection of documents (corpus).

Formula Breakdown:
TF (Term Frequency):
How frequently a word appears in a document.

javascript
Copy code
TF = (Number of times term t appears in a document) / (Total terms in the document)
IDF (Inverse Document Frequency):
How rare the word is across all documents.

ini
Copy code
IDF = log_e(Total number of documents / Number of documents containing the term)
TF-IDF = TF  IDF



```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    ngram_range=(1, 2),     # Use unigrams + bigrams
    min_df=2,               # Ignore rare terms
    max_df=0.9,             # Ignore too common terms
    stop_words='english',
    sublinear_tf=True       # Apply sublinear term frequency scaling
)
# Replace 'text_train' with 'df['review']' to use the 'review' column
x = vectorizer.fit_transform(df['review']).toarray()
```

Hyperparameter tunning

Hyperparameter tuning is the process of selecting the optimal set of hyperparameters for a machine learning model. Hyperparameters are values set before the learning process begins and control how the model is trained. These can significantly affect the models performance.

hyperparameter tuning using Grid Search cv

Grid Search CV (Cross-Validation) is a technique used for hyperparameter tuning by performing an exhaustive search over a specified hyperparameter grid. It evaluates model performance for each combination of hyperparameters and selects the best set based on cross-validation results.

Notes:
Cross-validation is crucial because it gives a better estimate of how your model will perform on unseen data.

Parallelization (n_jobs=-1) speeds up the search process by utilizing all CPU cores.

Be cautious about the search space. A large number of hyperparameters with a large range of values will significantly increase the time it takes to perform Grid Search.
